```python
import pandas as pd
from SE.snowflake import _get_DF, _get_sf_connection
from SE.credentials import SNOWFLAKE_ECON
```

```python
from tqdm.notebook import tqdm
```

```python
P = pd.date_range('2019-10-01', '2024-04-30', freq='W')
```

```python
S, E = P.to_period('W').start_time, P.to_period('W').end_time
```

```python
S[0], E[0]
```

```python
len(S)
```

```python
with open('./candidates-georef-v1.sql', 'r') as f:
    q = f.read()
```

```python
filter_params = dict(
    price_lim=(-0.3, 0.55),
    day_lim=(-90, 90),
    distance=1000
)
```

```python
SNOWFLAKE_ECON['warehouse'] = 'ITERABLE_2XL_WH'
```

```python
con = _get_sf_connection(**SNOWFLAKE_ECON)
```

```python
# S
```

```python
# results = []

for s, e in zip(S, E):
    print(s)
    Q = q.format(period_start=s, period_end=e, **filter_params)
    df = _get_DF(Q, con=con, fast=False)
    df.to_parquet(f'./data/candidates_georef_v4/{s:%Y-%m-%d}.parquet')
```
